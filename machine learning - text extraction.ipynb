```python
from PyPDF2 import PdfFileReader
```

```python
#Create Pdf File readr Object
pdf2 = PdfFileReader('C:\\Users\\srt\\Downloads\\Job Discription.pdf')
```

```python
page_2_object = pdf2.getPage(0)
```

```python
print(page_2_object)
```

```python
page_2_text = page_2_object.extractText()
```

```python
print(page_2_text)
```

```python
text = [page_1_text, page_2_text]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(text)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
print("\nSimilarity Scores:")
print(cosine_similarity(count_matrix))
```

```python
matchpercentage = cosine_similarity(count_matrix)[0][1]*100
matchpercentage = round(matchpercentage, 2) #round to two decimals
print("your resume matches about " + str(matchpercentage)+ "% of job description.")
```
